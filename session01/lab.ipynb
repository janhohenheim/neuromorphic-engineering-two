```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
headers = ['Vgs', 'Ids']
filename = "sweep-vgs.csv"
df = pd.read_csv(filename, names=headers, skiprows=1)

plt.rcParams['figure.figsize'] = [12, 9]
plt.rcParams.update({'font.size': 18})

plt.semilogy(df['Vgs'], df['Ids'])
plt.grid(True)
plt.show()
```
